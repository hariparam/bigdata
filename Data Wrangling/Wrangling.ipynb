```python
def jaccard_dist(a,b):
    return float(len(set(a).intersection(set(b))))/len(set(a).union(set(b)))
```

```python
jaccard_map={}
for r in routes_df:
     if routes_df[r].dtype !=pd.np.int64 and routes_df[r].dtype !=pd.np.float64:
        row_map={}
        for p in performance_s_df:
            if performance_s_df[p].dtype !=pd.np.int64 and performance_s_df[p].dtype !=pd.np.float64:
                x=jaccard_dist(routes_df[r].unique(),performance_s_df[p].unique())
                if x>0.0:
                    row_map[p]=x
        if len(row_map):
            jaccard_map[r]=row_map
```
