```python
import pandas as pd
from pandas import Series
```

```python
airports_df=pd.read_csv('airports.dat.txt',names=['airport_id','airport_name','airport_city','airport_country','airport_faa_code','airport_icao_code','airport_latitude','airport_longitude','airport_altitude','airport_tz_offset','airport_dst','airport_tz'])
airlines_df=pd.read_csv('airlines.dat.txt',names=['airline_id','airline_name','airline_alias','airline_iata_code','airline_icao_code','airline_callsign','airline_country','airline_active'])
routes_df=pd.read_csv('routes.dat.txt',names=['routes_airline_code','routes_airline_id','routes_source_airport_code','routes_source_airport_id','routes_destination_airport_code','routes_destination_airport_id','routes_codeshare','routes_stops_number','routes_equipment_code'])
performance_df=pd.read_csv('On_Time_On_Time_Performance_2016_10.csv')
```

```python
performance_df=performance_df[['Year','Month','DayofMonth','AirlineID','Carrier','FlightNum','Origin','Dest','ArrDelayMinutes','Cancelled']]
```

```python
# Replace NaNs with blanks if the column is a string, so we don’t get
# objects in the column
def fillna_col(series):
    if series.dtype is pd.np.dtype(object):
        return series.fillna('')
    else:
        return series
```

```python
#Nullify
def nullify(x):
    if x== '\\N':
        return pd.np.nan
    else:
        return x
```

```python
airlines_df=airlines_df.applymap(nullify)
airports_df=airports_df.applymap(nullify)
routes_df=routes_df.applymap(nullify)
```

```python
routes_df =routes_df.dropna(subset=['routes_airline_id','routes_source_airport_code','routes_source_airport_code','routes_source_airport_id','routes_destination_airport_code','routes_destination_airport_id'])
```

```python
airlines_df=airlines_df.apply(fillna_col)
airports_df=airports_df.apply(fillna_col)
routes_df=routes_df.apply(fillna_col)
performance_df=performance_df.apply(fillna_col)
```
