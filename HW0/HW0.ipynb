```python
import pyspark
sc=pyspark.SparkContext('local[*]')
rdd=sc.parallelize(range(1000))
rdd.takeSample(False,5)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

y=np.array(rdd.takeSample(False,100,1))
x=np.array(range(len(y)))

plt.title(str(len(y))+ ' random samples from RDD')
plt.xlabel('sample no')
plt.ylabel('Value')
plt.figtext(0.995,0.01,'CIS 700-003 student-harip', ha='right', va= 'bottom')
plt.scatter(x,y)
plt.tight_layout()
m,c=np.polyfit(x,y,1)
plt.plot(x,m*x+c)
plt.savefig('hw0.svg')
```
